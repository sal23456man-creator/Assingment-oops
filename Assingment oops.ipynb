```python
#  How is encapsulation achieved in Python..?

# In Python, encapsulation is achieved through a combination of programming conventions and name mangling
# 1. The Power of Convention (Single Underscore _)
# The primary way encapsulation is signaled in Python is by using a single underscore prefix (_) before an attribute or method name (e.g., _balance). 
# Convention: This is a clear signal to other programmers that the attribute or method is "intended for internal use only" and should be treated as private.
# Access: Python does not technically prevent external code from accessing or modifying _balance. The convention relies on programmer discipline


# 2. Name Mangling (Double Underscore __)
 # For a stronger form of encapsulation, Python uses a mechanism called name mangling, which employs a double underscore prefix (__) (e.g., __balance). 
#  Purpose: The Python interpreter automatically changes the name of the attribute internally to include the class name (e.g., __balance becomes _BankAccount__balance). This makes it significantly harder to access the attribute accidentally from outside the class.
#M echanism: This isn't true private access (you can still access it if you know the mangled name), but it effectively prevents naming conflicts in subclasses and discourages external access.


# 3. Using Getters and Setters (@property) 
# The most "Pythonic" way to implement controlled access (encapsulation) is by using properties. This allows you to define public methods for accessing and modifying data while keeping the data itself conventionally private (_ prefix). 
# This approach provides validation logic and computed attributes while maintaining a clean, simple syntax: 
```

```python
# What are class and static methods in Python..?

 Static Methods
# Static methods are functions logically grouped within a class (usually because they relate to the class's purpose) but do not interact with either the class's state or the object's instance state. They are defined using the @staticmethod decorator.
# Key Characteristics
# Decorator: Uses @staticmethod.
# No Special Parameters (self or cls): A static method takes no mandatory first arguments (self or cls). It behaves exactly like a regular function that simply happens to be defined inside the class namespace.
# Purpose: They are used for utility functions or helper functions that have a logical connection to the class but don't need any of its data


Class Methods
# Class methods are functions that operate on the class as a whole, rather than on an instance of the class. They are defined using the @classmethod decorator.
# Key Characteristics
# Decorator: Uses @classmethod.
# First Parameter (cls): The first parameter of a class method is conventionally named cls (short for class). Python automatically passes the class itself to this argument when the method is called.
# Purpose: They are primarily used for methods that need to access or modify the class's state (class variables) or to create alternative ways of instantiating the class (known as factory methods).
```

```python
#  What are the advantages of OOP..?

# Object-Oriented Programming (OOP) offers numerous advantages over traditional procedural programming by organizing software design around data (objects) rather than functions and logic.

# 1. Modularity and Reusability
# Modularity: Because objects manage their own state and behavior (encapsulation), changes in one part of the system have minimal impact on others. This simplifies troubleshooting and development by multiple teams.
# Code Reusability: Inheritance allows developers to create new classes based on existing ones, reusing proven logic and reducing the need to write redundant code from scratch. This saves significant time and effort.

# 2. Easier Maintenance and Debugging
# Isolation of Issues: When a bug occurs, it is usually contained within a specific object or class, rather than scattered across global functions. This isolation simplifies debugging and fixes.
# Centralized Logic: Common logic is stored in base classes (parent classes). Updating a single method in a parent class automatically updates the behavior for all child classes inheriting from it.


# 3. Improved Code Readability and Comprehension
# Real-World Modeling: Modeling entities as objects (like a Car object or a User object) makes the codebase more intuitive and accessible to new developers joining a project.
# Clear Interfaces: Abstraction and encapsulation ensure that users of a class only need to understand the simple public interface (what it does), not the complex internal implementation details (how it does it).


# 4.Flexibility (Polymorphism)
# Polymorphism allows objects of different classes to be treated through a single, common interface.
# Extensibility: This flexibility means you can easily add new functionality or new object types to a system without having to modify the existing core logic. The program can automatically handle new variations.
```

```python
# 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
#    and further derive a class ElectricCar that adds a battery attribute.

class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type
        
    def display_type(self):
        print(f"This is a {self.vehicle_type}.")


class Car(Vehicle):   
    def __init__(self, brand):
       
        super().__init__(vehicle_type="Car") 
        self.brand = brand
       

    def display_brand(self):
        print(f"Brand: {self.brand}")


class ElectricCar(Car):
    
    def __init__(self, brand, battery_kwh):
        
        super().__init__(brand) 
        self.battery_kwh = battery_kwh
        
    def display_battery(self):
        print(f"Battery Capacity: {self.battery_kwh} kWh")




print("Creating a standard Car object:")
my_car = Car(brand="BMW")
my_car.display_type()
my_car.display_brand()


print("Creating an Electric Car object (multilevel inheritance):")
my_electric_car = ElectricCar(brand="TATA Ev", battery_kwh=100)

my_electric_car.display_type()    
my_electric_car.display_brand()   
my_electric_car.display_battery() 
```

```python
# 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
#    Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("A generic bird is flying high in the sky.")

class Sparrow(Bird):
    def fly(self):
        print("The sparrow flits quickly between branches.")

class Penguin(Bird):
    def fly(self):
        print("The penguin cannot fly, but it waddles on land and swims quickly in water.")


birds = [Bird(), Sparrow(), Penguin()]

print("Demonstrating polymorphism:")

for bird in birds:
    bird.fly()
    
```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
#    balance and methods to deposit, withdraw, and check balance.

class BankAccount:
   
    def __init__(self, initial_balance=0:
        self.__balance = initial_balance 
        print(f"Account created with initial balance: ${self.__balance:.2f}")

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount:.2f}")
            return True
        else:
            print("Error: Deposit amount must be positive.")
            return False

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount:.2f}")
            return True
        elif amount > self.__balance:
            print("Error: Insufficient funds.")
            return False
        else:
            print("Error: Withdrawal amount must be positive.")
            return False

    @property
    def check_balance(self):
        print("Checking balance...")
        return self.__balance

my_account = BankAccount(initial_balance=1000)

print("-" * 30)


my_account.deposit(200)

my_account.withdraw(150)

my_account.withdraw(5000)

print(f"Current Balance: ${my_account.check_balance:.2f}")

print("-" * 30)

try:
    print(my_account.__balance)
except AttributeError as e:
    print(f"Attempted direct access to private attribute: {e}")

my_account.__balance = 9999999999
print(f"Balance after attempted direct modification: ${my_account.check_balance:.2f}") 
```

```python
# 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
#    and Piano that implement their own version of play().

class Instrument:
    def play(self):
        print("Playing a generic instrument sound.")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar strings.")

class Piano(Instrument):
    def play(self):
        print("Pressing the piano keys.")


def start_performance(instrument_object):
    instrument_object.play()

guitar = Guitar()
piano = Piano()
generic_instrument = Instrument()

print("Demonstrating runtime polymorphism:")
print("-" * 40)

start_performance(generic_instrument)
start_performance(guitar)
start_performance(piano)

print("-" * 40)

instruments = [Guitar(), Piano(), Instrument()]

for inst in instruments:
    inst.play()
    print("-" * 40)
```

```python
# 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
#    method subtract_numbers() to subtract two numbers.

class MathOperations:    
    @classmethod
    def add_numbers(cls, num1, num2):
        print(f"Using class method: Adding {num1} and {num2}")
        return num1 + num2

    @staticmethod
    def subtract_numbers(num1, num2):
        print(f"Using static method: Subtracting {num2} from {num1}")
        return num1 - num2


sum_result = MathOperations.add_numbers(10, 5)
print(f"Sum: {sum_result}")

print("-" * 30)

diff_result = MathOperations.subtract_numbers(10, 5)
print(f"Difference: {diff_result}")
```

```python
# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#    vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Vector(new_x, new_y)

vector_a = Vector(x=5, y=10)
vector_b = Vector(x=3, y=2)

print(f"Vector A: {vector_a}")
print(f"Vector B: {vector_b}")
print("-" * 30)

vector_c = vector_a + vector_b

print(f"Vector A + Vector B = Vector C")
print(f"Vector C: {vector_c}")

```

```python
# 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
#     the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades
        print(f"Student '{self.name}' initialized with grades: {self.grades}")

    def average_grade(self):
        if not self.grades:
            return 0.0
        
        
        total_sum = sum(self.grades)
        count = len(self.grades)
        return total_sum / count


student_a = Student(name="salman", grades=[90, 85, 92, 78, 95])
student_b = Student(name="arbaj", grades=[75, 88, 81])
student_c = Student(name="rahul", grades=[]) 

print("-" * 30)


average_a = student_a.average_grade()
print(f"{student_a.name}'s average grade: {average_a:.2f}")


average_b = student_b.average_grade()
print(f"{student_b.name}'s average grade: {average_b:.2f}")


average_c = student_c.average_grade()
print(f"{student_c.name}'s average grade: {average_c:.2f}")
```

```python
# 13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the
#     area.

class Rectangle:
    def __init__(self, width=0, height=0):
        self.width = width
        self.height = height
        print(f"Rectangle initialized with dimensions: W={self.width}, H={self.height}")

    def set_dimensions(self, width, height):
        if width > 0 and height > 0:
            self.width = width
            self.height = height
            print(f"Dimensions updated to: W={self.width}, H={self.height}")
        else:
            print("Error: Dimensions must be positive values.")

    def area(self):
        if self.width <= 0 or self.height <= 0:
            print("Cannot calculate area with non-positive dimensions.")
            return 0
        
        return self.width * self.height

rect1 = Rectangle() 

rect1.set_dimensions(width=10, height=5)


area1 = rect1.area()
print(f"The area of the rectangle is: {area1}") 

print("-" * 30)


rect2 = Rectangle(width=6, height=3)
area2 = rect2.area()
print(f"The area of the second rectangle is: {area2}") 

print("-" * 30)


rect1.set_dimensions(width=-2, height=5)
print(f"The area of rect1 is still: {rect1.area()}")
```

```python
# 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#     and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        base_salary = self.hourly_rate * self.hours_worked
        return base_salary

class Manager(Employee):
    def __init__(self, name, hourly_rate, hours_worked, bonus):
        super().__init__(name, hourly_rate, hours_worked)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        total_salary = base_salary + self.bonus
        return total_salary


emp1 = Employee("salman", hourly_rate=20, hours_worked=160)
salary_emp1 = emp1.calculate_salary()
print(f"{emp1.name}'s Salary: ${salary_emp1:.2f}") # Output: 3200.00

print("-" * 30)


manager1 = Manager("arbaj", hourly_rate=20, hours_worked=160, bonus=500)
salary_manager1 = manager1.calculate_salary()
print(f"{manager1.name}'s Salary: ${salary_manager1:.2f}") 
```

```python
# 15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
#     calculates the total price of the product.

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        
        if price >= 0 and quantity >= 0:
            self.price = price
            self.quantity = quantity
        else:
            raise ValueError("Price and quantity cannot be negative.")
        
        print(f"Product '{self.name}' created.")

    def total_price(self):
        total = self.price * self.quantity
        return total

product1 = Product(name="Laptop", price=999.99, quantity=5)
product2 = Product(name="Mouse", price=25.50, quantity=50)

print("-" * 30)


total1 = product1.total_price()
print(f"Total price for {product1.name}: ${total1:.2f}")

total2 = product2.total_price()
print(f"Total price for {product2.name}: ${total2:.2f}")

print("-" * 30)


product1.quantity = 10
total1_updated = product1.total_price()
print(f"Updated total price for {product1.name} (Quantity 10): ${total1_updated:.2f}")
```

```python
# 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
#    implement the sound() method.?


class Animal:
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def sound(self):
        pass 

    def display_name(self):
        print(f"I am a {self.__class__.__name__} named {self.name}.")

class Cow(Animal):
    def sound(self):
        return "Moooooo..!"

class Sheep(Animal):
    def sound(self):
        return "Baaabaa...!"

cow1 = Cow(name="Betsy")
sheep1 = Sheep(name="Dolly")


print("Demonstrating animal sounds:")
print("-" * 30)


farm_animals = [cow1, sheep1]

for animal in farm_animals:
    animal.display_name()
    print(f"The sound I make is: {animal.sound()}")
    print("-" * 30)
```
